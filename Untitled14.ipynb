```python
# Split data into train and test sets
x_train1 = data[:15000, 0:7]
x_test1=data[15000:,0:7]
y_train = data[:15000, 7]
y_test = data[15000:, 7]


# Add bias term
x_train = np.ones((x_train1.shape[0], x_train1.shape[1] + 1))
x_train[:, 1:] = x_train1


x_test = np.ones((x_test1.shape[0], x_test1.shape[1] + 1))
x_test[:, 1:] = x_test1


# Standardize features
scaler = StandardScaler()
scaler.fit(x_train[:, 1:])
x_train[:, 1:] = scaler.transform(x_train[:, 1:])
x_test[:, 1:] = scaler.transform(x_test[:, 1:])

# Initialize theta with random values
theta = np.random.uniform(0, 1, size=(x_train.shape[1], 1))

# Set hyperparameters
learning_rate = 0.01
num_iterations = 1000
```

```python
# Hyperparameters
n_iterations = 1000
learning_rate = 0.01

# Number of training examples (m) and features (n)
m = x_train.shape[0]
n = x_train.shape[1]

# Reshape y_train to be a column vector for correct broadcasting
y_train_reshaped = y_train.reshape(-1, 1)

# Initialize a list to store cost history
cost_history = []

# Gradient Descent
for iteration in range(n_iterations):
  # Calculate predictions
  predictions = np.dot(x_train, theta)

  # Calculate errors
  # This will now be (m, 1) after reshaping y_train
  errors = predictions - y_train_reshaped

  # Calculate cost (Mean Squared Error)
  # Using numpy's vectorized operations for efficiency and correctness
  cost = np.sum(errors**2) / (2 * m)
  cost_history.append(cost)

  # Calculate gradients
  # Gradient is (1/m) * X_transpose * errors
  # x_train.T is (n, m), errors is (m, 1) -> gradient_term is (n, 1)
  gradient_term = np.dot(x_train.T, errors) / m
  # theta is (n, 1), gradient_term is (n, 1)
  theta -= learning_rate * gradient_term
```

```python
iteration = np.arange(1, n_iterations + 1)
plt.plot(iteration, cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

### Evaluation of SGDRegressor

```python

# Load data from the given URL
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# Remove rows with missing values
df = raw_df.dropna()

# Extract features (x) and target (y)
data = np.array(df)
x = data[:, 0:10]
y = data[:, 10]

# Standardize the features
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.8, shuffle=True)

# Create and train the SGDRegressor model
model = SGDRegressor(max_iter=1000, alpha=0.01)
model.fit(x_train, y_train)

# Get the model parameters (coefficients)
parameters = model.coef_
print("Model Coefficients (Theta_1 to Theta_10):", parameters)

# Get the intercept (Theta_0)
theta0 = model.intercept_
print("Intercept (Theta_0):", theta0)

# Make predictions on the test data
predictions = model.predict(x_test)

# Calculate and print the mean absolute error
absolute_error = mean_absolute_error(y_test, predictions)
print("Mean Absolute Error (MAE):", absolute_error)

# Calculate and print the mean squared error
mean_error = mean_squared_error(y_test, predictions)
print("Mean Squared Error (MSE):", mean_error)
```

###Linear Regression from Scratch with L1 & L2 Regularization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

housing = datasets.fetch_california_housing()

X = housing.data
y = housing.target

df = pd.DataFrame(X, columns=housing.feature_names)
df["Target"] = y

print(df.head())


# ==========================================
# TRAIN TEST SPLIT
# ==========================================

X_train_raw = X[:15000]
X_test_raw = X[15000:]

y_train = y[:15000]
y_test = y[15000:]


# ==========================================
# ADD BIAS COLUMN
# ==========================================

X_train = np.ones((X_train_raw.shape[0], X_train_raw.shape[1] + 1))
X_train[:, 1:] = X_train_raw

X_test = np.ones((X_test_raw.shape[0], X_test_raw.shape[1] + 1))
X_test[:, 1:] = X_test_raw


# ==========================================
# STANDARDIZE FEATURES
# ==========================================

scaler = StandardScaler()

X_train[:, 1:] = scaler.fit_transform(X_train[:, 1:])
X_test[:, 1:] = scaler.transform(X_test[:, 1:])


# ==========================================
# INITIALIZE PARAMETERS
# ==========================================

np.random.seed(42)

theta = np.random.randn(X_train.shape[1], 1)

learning_rate = 0.01
n_iterations = 1000

# Regularization parameters
lambda_l1 = 0.01
lambda_l2 = 0.01

m = X_train.shape[0]

# Convert y to column vector
y_train = y_train.reshape(-1, 1)


# ==========================================
# TRAINING USING GRADIENT DESCENT
# ==========================================

cost_history = []

for iteration in range(n_iterations):

    # --------------------------------------
    # FORWARD PROPAGATION
    # --------------------------------------

    predictions = np.dot(X_train, theta)

    errors = predictions - y_train

    # --------------------------------------
    # COST FUNCTION
    # --------------------------------------

    mse_cost = np.sum(errors ** 2) / (2 * m)

    l1_cost = (lambda_l1 / m) * np.sum(np.abs(theta[1:]))

    l2_cost = (lambda_l2 / (2 * m)) * np.sum(theta[1:] ** 2)

    total_cost = mse_cost + l1_cost + l2_cost

    cost_history.append(total_cost)

    # --------------------------------------
    # GRADIENTS
    # --------------------------------------

    gradients = np.dot(X_train.T, errors) / m

    # L1 Regularization Gradient
    gradients[1:] += (lambda_l1 / m) * np.sign(theta[1:])

    # L2 Regularization Gradient
    gradients[1:] += (lambda_l2 / m) * theta[1:]

    # --------------------------------------
    # UPDATE PARAMETERS
    # --------------------------------------

    theta -= learning_rate * gradients


# ==========================================
# PLOT COST FUNCTION
# ==========================================

iterations = np.arange(1, n_iterations + 1)

plt.figure(figsize=(10, 5))

plt.plot(iterations, cost_history)

plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Gradient Descent Convergence")

plt.grid(True)

plt.show()


# ==========================================
# MODEL PARAMETERS
# ==========================================

print("\nTheta Values:\n")
print(theta)


# ==========================================
# PREDICTIONS
# ==========================================

predictions = np.dot(X_test, theta)

predictions = predictions.flatten()


# ==========================================
# EVALUATION METRICS
# ==========================================

MAE = mean_absolute_error(y_test, predictions)

MSE = mean_squared_error(y_test, predictions)

RMSE = np.sqrt(MSE)

print("\n========== MODEL PERFORMANCE ==========")

print(f"MAE  : {MAE:.4f}")
print(f"MSE  : {MSE:.4f}")
print(f"RMSE : {RMSE:.4f}")


# ==========================================
# SAMPLE PREDICTIONS
# ==========================================

results = pd.DataFrame({
    "Actual": y_test[:10],
    "Predicted": predictions[:10]
})

print("\nSample Predictions:\n")
print(results)
```
